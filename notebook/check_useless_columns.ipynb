```python
from pprint import pprint

import pandas as pd

from consts import REPO_ROOT
```

```python
df = pd.read_csv(REPO_ROOT / "data/raw/train.csv")
df.head()
```

```python
# columns with all null values
null_columns = df.columns[df.isnull().all()].tolist()

print(len(null_columns))
pprint(null_columns)
```

```python
# columns with constant values
constant_columns = df.columns[df.nunique() == 1].tolist()

print(len(constant_columns))
pprint(constant_columns)
```

```python
# duplicated columns

# detect duplicated columns
columns_filtered = [
    c for c in df.columns if c not in set(null_columns + constant_columns)
]

duplicate_column_groups: list[list[str]] = []

for col in columns_filtered:
    found_match = False
    for group in duplicate_column_groups:
        if df[col].equals(df[group[0]]):
            group.append(col)
            found_match = True
            break
    if not found_match:
        duplicate_column_groups.append([col])

duplicate_column_groups = [
    group for group in duplicate_column_groups if len(group) >= 2
]

print(len(duplicate_column_groups))
pprint(duplicate_column_groups)
```

```python
# duplicated columns to remove
duplicate_columns = [
    column for group in duplicate_column_groups for column in group[1:]
]

print(len(duplicate_columns))
pprint(duplicate_columns)
```
